```python
#Change the center pixel of the image
import cv2
import matplotlib.pyplot as plt

# Read an image in color (not grayscale)
image_path = r'C:\Users\DELL\Downloads\Rose.jpg'
image_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)  # Load as a color image

# Check if the image was loaded successfully
if image_bgr is None:
    raise ValueError("Image not loaded. Check the file path and format.")

# Convert the BGR image to RGB format for display in matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Get the dimensions of the image
height, width, _ = image_rgb.shape

# Compute the coordinates of the center pixel
center_y, center_x = height // 2, width // 2

# Define the size of the block to change (e.g., a fraction of the image dimensions)
block_height = height // 30  # Change to a fraction of the height
block_width = width // 30  # Change to a fraction of the width

# Ensure the block size is odd to center the change on the exact pixel
block_height = block_height if block_height % 2 == 1 else block_height + 1
block_width = block_width if block_width % 2 == 1 else block_width + 1

# Define the block of pixels around the center
start_y = max(center_y - block_height // 2, 0)
end_y = min(center_y + block_height // 2 + 1, height)
start_x = max(center_x - block_width // 2, 0)
end_x = min(center_x + block_width // 2 + 1, width)

# Change a block of pixels around the center pixel to red (255, 0, 0)
image_rgb[start_y:end_y, start_x:end_x] = [255, 0, 0]  # Red

# Save the modified RGB image to a new file
saved_image_path = r'C:\Users\DELL\Downloads\Photo_pixel.jpg'
cv2.imwrite(saved_image_path, cv2.cvtColor(image_rgb, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving

# Display the modified RGB image
plt.figure(figsize=(7.5, 5))
plt.imshow(image_rgb)
plt.title('Modified RGB Image with Center Pixel Block Changed')
plt.axis('off')
plt.show()
```

```python
#Write the text Am proud to be an Indian on the Image 
from PIL import Image, ImageDraw, ImageFont

# Load your image
image_path = r'C:\Users\DELL\Downloads\Rose.jpg'
image = Image.open(image_path)

# Convert image to RGB (if not already in that mode)
image = image.convert('RGB')

# Define the color blue
blue = (0, 0, 255)

# Modify the top left 20x20 pixel region
for x in range(20):
    for y in range(20):
        image.putpixel((x, y), blue)

# Initialize ImageDraw
draw = ImageDraw.Draw(image)

# Define text and font
text = "I am proud to be an Indian"
font_size = 30

# Path to a TTF font file, you might need to download one if you don't have it
# On Windows, you can use a system font like Arial
font_path = "C:\\Windows\\Fonts\\arial.ttf"
font = ImageFont.truetype(font_path, font_size)

# Define text position and color
text_position = (100, 50)  # Adjust position as needed
text_color = (0, 0, 0)  # White color

# Add text to image
draw.text(text_position, text, fill=text_color, font=font)

# Save or display the modified image
image.show()
image.save(r'C:\Users\DELL\Downloads\Modified_Rose_with_text.jpg')
```

```python
#Draw circle on the and write I am prou to be an Indian on the within the circle
import cv2

# Load your image
image_path = r'C:\Users\DELL\Downloads\Rose.jpg'
image = cv2.imread(image_path)

# Define the center and radius of the circle
center = (image.shape[1] // 2, image.shape[0] // 2)  # Center of the image
radius = 200  # Adjust the radius as needed

# Draw the circle
circle_color = (255, 0, 0)  # Blue color in BGR
cv2.circle(image, center, radius, circle_color, thickness=5)  # Thickness of 5

# Define the text to be written inside the circle
text = "I am proud to be an Indian"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_thickness = 2
text_color = (0, 0, 0)  # White color in BGR

# Calculate the size of the text box
(text_width, text_height), baseline = cv2.getTextSize(text, font, font_scale, font_thickness)

# Calculate the position to center the text within the circle
text_x = center[0] - text_width // 2
text_y = center[1] + text_height // 2

# Put the text on the image
cv2.putText(image, text, (text_x, text_y), font, font_scale, text_color, font_thickness, lineType=cv2.LINE_AA)

# Save or display the modified image
cv2.imshow('Modified Image', image)
cv2.imwrite(r'C:\Users\DELL\Downloads\Modified_Rose_with_circle_and_text.jpg', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Convrt the RGB color modules to other modules(cmy suffix,ycbc suffix r)
import cv2
import numpy as np

# Load your image
image_path = r'C:\Users\DELL\Downloads\Rose.jpg'
image = cv2.imread(image_path)

# Normalize the RGB values to [0, 1] by dividing by 255
rgb_normalized = image / 255.0

# Convert to CMY
cmy = 1 - rgb_normalized

# Convert back to [0, 255] range for display
cmy = (cmy * 255).astype(np.uint8)

# Save or display the CMY image
cv2.imshow('CMY Image', cmy)
cv2.imwrite(r'C:\Users\DELL\Downloads\CMY_Image.jpg', cmy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
